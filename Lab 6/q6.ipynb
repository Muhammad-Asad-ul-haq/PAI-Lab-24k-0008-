```python
import pandas as pd

df = pd.read_csv("Software houses PK .csv")
df
```

```python
city_counts = df["City"].value_counts()
print(city_counts)
```

```python
services_expanded = (
    df["Services"]
    .str.lower()
    .str.replace(",", " ")
    .str.split()
    .explode()
)

industry_counts = services_expanded.value_counts()

print(industry_counts)
```

```python
tech_words = (
    df["Services"]
    .str.lower()
    .str.replace(",", " ")
    .str.split()
    .explode()
)

technology_counts = tech_words.value_counts()

print(technology_counts)
```
